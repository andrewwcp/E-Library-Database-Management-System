```python
def generate_libraries_data(num, lib_name):
    """
    Generate data for libraries.

    Args:
        num (int): Number of rows of data to generate.
        lib_name (list): List of library names.

    Returns:
        pandas.DataFrame: A DataFrame containing library data with columns:
            - library_id (int): Unique identifier for each library.
            - library_name (str): Name of the library.
            - email (str): Email address of the library (generated based on library name).
            - contact_number (str): Contact phone number of the library (generated).
    """
    library_names = lib_name
    email = [f"{library_name.replace(' ', '_')}@umail.com" for library_name in library_names]

    library = {"library_id": [i+1 for i in range(num)],
               "library_name": [i for i in library_names],
               "email": [i for i in email],
               "contact_number": []}

    for i in range(num):
      phone_number = fake.phone_number()
      library['contact_number'].append(phone_number)

    df = pd.DataFrame(library)
    return df

library_names = ["Perpustakaan Universitas Raharja", "Grahatama Pustaka", "Pustakalana", "Library Satu University", "Perpustakaan Universitas Tama"]
df_library = pd.DataFrame(generate_libraries_data(num=5, lib_name=library_names))
df_library
```

```python
def generate_users_data(num=5):
    """
    Generate user data.

    Args:
        num (int, optional): Number of user records to generate. Defaults to 5.

    Returns:
        pandas.DataFrame: DataFrame containing user data with columns:
            - user_id (int): Unique identifier for each user.
            - username (str): Username generated based on first name, last name, and phone number.
            - first_name (str): User's first name.
            - last_name (str): User's last name.
            - email (str): User's email address.
            - password (str): User's password.
            - contact_number (str): User's contact phone number.
    """

    first_name_list = []
    last_name_list = []
    phone_number_list = []
    password_list = []
    username_list = []
    email_list = []

    for i in range(num):
      first_name = fake.first_name()
      last_name = fake.last_name()
      phone_number = fake.phone_number()
      password = fake.password()
      username = f"{first_name.lower()}_{last_name.lower()}{phone_number[-2:-1]}"
      email = f"{first_name.lower()}.{last_name.lower()}@usermail.com"

      first_name_list.append(first_name)
      last_name_list.append(last_name)
      phone_number_list.append(phone_number)
      password_list.append(password)
      username_list.append(username)
      email_list.append(email)

    users = {"user_id": [i+1 for i in range(num)],
              "username": username_list,
              "first_name": first_name_list,
              "last_name": last_name_list,
              "email": email_list,
              "password": password_list,
              "contact_number": phone_number_list
             }

    df = pd.DataFrame(users)
    return df

df_users = pd.DataFrame(generate_users_data(num=20))
df_users.head()
```

```python
def generate_authors_data(num, books_metadata):
    """
    Generate author data.

    Args:
        num (int): Number of author records to generate.
        books_metadata (pandas.DataFrame): DataFrame containing book metadata.

    Returns:
        pandas.DataFrame: DataFrame containing author data with columns:
            - author_id (int): Unique identifier for each author.
            - name (str): Author's name.
            - email (str): Author's email address (generated based on name).
            - contact_number (str): Author's contact phone number (generated).

    """
    author_name_list = books_metadata['Book-Author'].unique().tolist()
    email_list = [f"{name}.{str(index+12)}@authormail.com" for index, name in enumerate(author_name_list)]
    email_list = [email.split(' ')[-1] for email in email_list]
    phone_number_list = []

    for i in range(num):
      phone_number = fake.phone_number()
      phone_number_list.append(phone_number)


    author = {"author_id": [i+1 for i in range(num)],
               "name": author_name_list[:num],
               "email": email_list[:num],
               "contact_number": phone_number_list}

    df = pd.DataFrame(author)
    return df

df_authors = pd.DataFrame(generate_authors_data(100, books_metadata))
df_authors.head()
```

```python
def generate_publishers_data(num, books_metadata):
    """
    Generate publisher data.

    Args:
        num (int): Number of publisher records to generate.
        books_metadata (pandas.DataFrame): DataFrame containing book metadata.

    Returns:
        pandas.DataFrame: DataFrame containing publisher data with columns:
            - publisher_id (int): Unique identifier for each publisher.
            - name (str): Publisher's name.
            - address (str): Publisher's address.
            - email (str): Publisher's email address (generated based on name).
            - contact_number (str): Publisher's contact phone number (generated).

    """

    publisher_name_list = books_metadata['Publisher'].unique().tolist()
    address_list = []
    email_list = [f"{name}.{str(index+12)}@publishermail.com" for index, name in enumerate(publisher_name_list)]
    email_list = [email.split(' ')[-1] for email in email_list]
    phone_number_list = []

    for i in range(num):
      phone_number = fake.phone_number()
      address = fake.address()
      phone_number_list.append(phone_number)
      address_list.append(address)

    author = {"publisher_id": [i+1 for i in range(num)],
               "name": publisher_name_list[:num],
               "address": address_list,
               "email": email_list[:num],
               "contact_number": phone_number_list}

    df = pd.DataFrame(author)
    return df

df_publishers = pd.DataFrame(generate_publishers_data(70, books_metadata))
df_publishers.head()
```

```python
def generate_categories_data(num, cat_name):
    """
    Generate category data.

    Args:
        num (int): Number of category records to generate.
        cat_name (list): List of category names.

    Returns:
        pandas.DataFrame: DataFrame containing category data with columns:
            - category_id (int): Unique identifier for each category.
            - category_name (str): Name of the category.

    """
    category_name = cat_name
    category = {"category_id": [i+1 for i in range(num)],
               "category_name": [i for i in category_name]}

    df = pd.DataFrame(category)
    return df

category_name = ["Mystery", "Thriller", "Science Fiction", "Fantasy", "Romance", "Non-Fiction", "Biography"]
df_categories = pd.DataFrame(generate_categories_data(num=7, cat_name=category_name))
df_categories
```

```python
def generate_books_data(num, books_metadata, df_categories, df_authors, df_publishers):
    """
    Generate book data.

    Args:
        num (int): Number of book records to generate.
        books_metadata (pandas.DataFrame): DataFrame containing book metadata.
        df_categories (pandas.DataFrame): DataFrame containing category data.
        df_authors (pandas.DataFrame): DataFrame containing author data.
        df_publishers (pandas.DataFrame): DataFrame containing publisher data.

    Returns:
        pandas.DataFrame: DataFrame containing book data with columns:
            - book_id (int): Unique identifier for each book.
            - book_title (str): Title of the book (cleaned from non-alphanumeric characters).
            - date_of_publication (datetime.date): Date of publication of the book.
            - category_id (int): ID of the category the book belongs to.
            - author_id (int): ID of the author of the book.
            - publisher_id (int): ID of the publisher of the book.

    """

    def clean_string(input_string):
        # Remove non-alphanumeric characters
        cleaned_string = re.sub(r'[^a-zA-Z0-9\s]', '', input_string)
        return cleaned_string

    # Assuming you have a DataFrame books_metadata with a column 'Book-Title'
    book_title_list = books_metadata['Book-Title'].unique().tolist()

    # Clean the book titles using the clean_string function
    cleaned_titles = [clean_string(title) for title in book_title_list]

    date_of_pub_list = []
    category_id_list = []
    author_id_list = []
    publisher_id_list = []

    for i in range(num):
      date_of_pub = fake.date_time_between(start_date=datetime(2000, 1, 1), end_date=datetime(2023, 1, 1)).date()
      date_of_pub_list.append(date_of_pub)
      category_id_list.append(random.choice(df_categories['category_id']))
      author_id_list.append(random.choice(df_authors['author_id']))
      publisher_id_list.append(random.choice(df_publishers['publisher_id']))

    author = {"book_id": [i+1 for i in range(num)],
               "book_title": cleaned_titles[:num],
               "date_of_publication": date_of_pub_list,
               "category_id": category_id_list,
               "author_id": author_id_list,
               "publisher_id": publisher_id_list}

    df = pd.DataFrame(author)
    return df

df_books = pd.DataFrame(generate_books_data(1000, books_metadata, df_categories, df_authors, df_publishers))
df_books.head()
```

```python
def generate_bookatlibrary_data(num, df_library):
    """
    Generate data for books available at libraries.

    Args:
        num (int): Number of book-at-library records to generate.
        df_library (pandas.DataFrame): DataFrame containing library data.

    Returns:
        pandas.DataFrame: DataFrame containing book-at-library data with columns:
            - library_book_id (int): Unique identifier for each library-book entry.
            - library_id (int): ID of the library where the book is available.
            - book_id (int): ID of the book available at the library.
            - quantity (int): Number of copies of the book available at the library.

    """

    book_at_library = {"library_book_id": [i+1 for i in range(num)],
                       "library_id": [random.choice(df_library['library_id']) for _ in range(num)],
                       "book_id": [i+1 for i in range(num)],
                       "quantity": [random.randint(1, 4) for _ in range(num)]}

    df = pd.DataFrame(book_at_library)
    return df

df_book_at_library = pd.DataFrame(generate_bookatlibrary_data(1000, df_library))
df_book_at_library = df_book_at_library.drop_duplicates(subset=['library_id', 'book_id'], keep='last')
df_book_at_library
```

```python
def generate_loan_data(num, df_users, df_book_at_library):
    """
    Generate data for loans.

    Args:
        num (int): Number of loan records to generate.
        df_users (pandas.DataFrame): DataFrame containing user data.
        df_book_at_library (pandas.DataFrame): DataFrame containing book-at-library data.

    Returns:
        pandas.DataFrame: DataFrame containing loan data with columns:
            - loan_id (int): Unique identifier for each loan.
            - user_id (int): ID of the user taking the loan.
            - library_book_id (int): ID of the book-at-library entry for the loan.
            - loan_date (datetime): Date when the loan was issued.
            - due_date (datetime): Date when the loan is due.
            - status (str): Status of the loan ('On Loan', 'Returned', 'Overdue').
            - return_date (datetime): Date when the book was returned, if applicable.

    """

    loan_date_list = []
    due_date_list = []
    return_date_list = []

    for i in range(num):
        loan_date = fake.date_time_between(start_date=datetime(2024, 1, 1), end_date=datetime.now())
        due_date = loan_date + timedelta(days=14)
        loan_date_list.append(loan_date)
        due_date_list.append(due_date)
        return_date_list.append(None)

    status_list = random.choices(['On Loan', 'Returned', 'Overdue'], weights=[0.3, 0.6, 0.2], k=num)
    for i, status in enumerate(status_list):
        if status == 'Returned':
             return_date_list[i] = loan_date_list[i] + timedelta(days=random.randint(1, 13))
        elif status == 'Overdue':
             return_date_list[i] = loan_date_list[i] + timedelta(days=random.randint(15, 20))
        elif status == 'On Loan':
             return_date_list[i] = None

    loan_data = {
        "loan_id": [i + 1 for i in range(num)],
        "user_id": [random.choice(df_users['user_id']) for _ in range(num)],
        "library_book_id": [random.choice(df_book_at_library['library_book_id']) for _ in range(num)],
        "loan_date": loan_date_list,
        "due_date": due_date_list,
        "status": status_list,
        "return_date": return_date_list,
    }

    df = pd.DataFrame(loan_data)
    return df


df_loan = pd.DataFrame(generate_loan_data(1500, df_users, df_book_at_library))
df_loan.head(30)
```

```python
def generate_hold_data(num, df_users, df_book_at_library):
    """
    Generate data for holds.

    Args:
        num (int): Number of hold records to generate.
        df_users (pandas.DataFrame): DataFrame containing user data.
        df_book_at_library (pandas.DataFrame): DataFrame containing book-at-library data.

    Returns:
        pandas.DataFrame: DataFrame containing hold data with columns:
            - hold_id (int): Unique identifier for each hold.
            - user_id (int): ID of the user placing the hold.
            - library_book_id (int): ID of the book-at-library entry for the hold.
            - hold_date (datetime): Date when the hold was placed.
            - expire_date (datetime): Date when the hold expires, or None if still active.
            - status (str): Status of the hold ('In Queue' or 'Expired').

    """
    hold_start_date_list = []
    hold_end_date_list = []
    expire_date_list = []
    status_list = []

    for i in range(num):
        hold_start_date = fake.date_time_between(start_date=datetime(2024, 4, 25), end_date=datetime.now())
        hold_start_date_list.append(hold_start_date)

        if hold_start_date >= datetime(2024, 5, 2) and hold_start_date < datetime.now():
          expire_date = None
        else:
          expire_date = hold_start_date + timedelta(days=7)
        expire_date_list.append(expire_date)

        if expire_date is None:
          status = 'In Queue'
        else:
          status = 'Expired'
        status_list.append(status)

    hold_data = {
        "hold_id": [i + 1 for i in range(num)],
        "user_id": [random.choice(df_users['user_id']) for _ in range(num)],
        "library_book_id": [random.choice(df_book_at_library['library_book_id']) for _ in range(num)],
        "hold_date": hold_start_date_list,
        "expire_date": expire_date_list,
        "status": status_list,
    }

    df = pd.DataFrame(hold_data)
    return df


df_hold = pd.DataFrame(generate_hold_data(600, df_users, df_book_at_library))
df_hold.head(30)
```
